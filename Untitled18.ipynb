```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import seaborn as sns
```

```python
data = pd.read_csv('/content/tsla_raw_data.csv')
data
```

```python
data.isnull().sum()
```

```python
data.fillna(0)
# Convert the 'Date' column to datetime data type
data['date'] = pd.to_datetime(data['date'])
```

```python
data.describe()
```

```python
def show(data: pd.DataFrame):
    df = data.copy()
    df = df.style.format(precision=2)
    df = df.background_gradient(cmap='Reds', axis=0)
    display(df)

# Pulling stock data
adj_date = data[data['date'] < '2023-12-31'].index
adj_date_fin = data.drop(adj_date)

df = pd.DataFrame(data, columns=['date','open', 'high', 'low', 'close', 'volume'])
df.set_index('date', inplace=True)

# Drop rows with missing data (NaN values)
df = df.dropna()

# Change the date format to '%Y-%m-%d'
df.index = df.index.strftime('%Y-%m-%d')

show(df.tail(10))
```

```python
adj_date_fin['date'] = pd.to_datetime(adj_date_fin['date'])

# Adjust the index display format
adj_date_fin['date'] = adj_date_fin['date'].dt.strftime('%d-%m')

plt.figure(figsize=(30,6))
plt.plot(adj_date_fin['date'], adj_date_fin['close'])
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(10))
plt.xlabel('Date')
plt.ylabel('Closing Price in USD')
plt.title('Tesla Stock Price in 2024')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.bar(adj_date_fin['date'], adj_date_fin['volume'])
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(10))
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Tesla Volume in 2024')
plt.grid()
plt.show()
```

```python
adj_date_fin['votality'] = (data['close'] / data['close'].shift(1)) - 1

# Create the bar plot for votality
plt.figure(figsize=(20, 8))
plt.bar(adj_date_fin['date'], adj_date_fin['votality'])
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Volatility of Tesla Volume in 2024')
plt.xlabel('date')
plt.ylabel('votality')
plt.grid()
plt.show()

# Create the histogram for votality
plt.figure(figsize=(10, 6))
adj_date_fin['votality'].hist(bins=5)
plt.title('Volatility of Tesla Volume in 2024')
plt.xlabel('votality')
plt.ylabel('Frequency')
plt.show()
```
